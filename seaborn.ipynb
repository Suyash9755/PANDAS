```python
# titanic data 
tit= sns.load_dataset('titanic')
tit.head()
```

```python
sns.countplot(x='alive',data=tit)
```

```python
sns.countplot(x='pclass',data=tit)
```

```python
# bar chart
sns.barplot(x='sex',y='class',hue='alive',data=tit)
```

```python
#important corraltion and heatmap
corra= tit.corr()
plt.Figure(figsize=[20,20])
sns.heatmap(corra,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'sizw':10},cmap='blues')
```
