```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
print(x,y,z)
```

```python
# ploting sin data into graph
plt.plot(x,y)
plt.show()
```

```python
# ploting cos data into graph
plt.plot(x,z)
plt.show()
```

```python
# adding titles and x-axis & y-axis lables 
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value') 
plt.title('sin wave')
plt.show()
```

```python
# parabola curve 
x = np.linspace(-20,20,40)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# adding the symbols in curve
plt.plot(x,y,'r.')   # red colour represent the colour and dot is ploting
plt.show()
```

```python
x = np.linspace(-10,10,100)
plt.plot(x, np.sin(x),'r-')
plt.plot(x , np.cos(x),'y--' )
plt.show()
```

```python
# bar plot 
fig = plt.figure()
ba = fig.add_axes([0,0,1,1])
languages = ['hindi','english', 'german','latin','french','russian']
peoples = [250,100,150,40,80,160]
ba.bar(languages,peoples, color='green')
plt.xlabel('languages')
plt.ylabel('peoples')
plt.show()
```

```python
# pie chart

fig = plt.figure()
ba = fig.add_axes([0,0,1,1])
languages = ['hindi','english', 'german','latin','french','russian']
peoples = [250,100,150,40,80,160]
ba.pie(peoples,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
# scatter plots 
x = np.linspace(0,10,20)
y= np.sin(x)
z= np.cos(x)
fig = plt.figure()
sx = fig.add_axes([0,0,1,1])
plt.scatter(x,y,color ='r')
plt.scatter(x,z,color='g')
plt.show()
```

```python
# 3d scatter chart
import matplotlib.pylab as plt
import numpy as np
fig = plt.figure()
sx = plt.axes(projection='3d')
x =40* np.random.random(100)
y = np.sin(x)
z=np.cos(x)
sx.scatter(y,z,x,c=x)
plt.show()
```
